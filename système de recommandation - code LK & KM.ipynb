## Linked List and Stack Implementation 

```python

class Node:
    def __init__(self, data):
        """
        Initializes a new instance of the Node class.

        Parameters:
        data: The data to be stored in the node.
        """
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        """
        Initializes a new instance of the LinkedList class.
        """
        self.head = None

    def append(self, data):
        """
        Appends a new node with the provided data to the end of the linked list.

        Parameters:
        data: The data to be stored in the new node.
        
        Returns:
        None
        """
         # Create a new node with the provided data : 
        new_node = Node(data)
        
        if not self.head: # If the linked list is empty
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def extend(self, my_list):
        """
        Extends the current linked list by appending all nodes from another linked list.

        Parameters:
        my_list: The linked list to extend the current list with.
        
        Returns:
        None
        """
        current = my_list.head
        while current:
            self.append(current.data)
            current = current.next

    def remove(self, data):
        """
        Removes the first occurrence of a node with the provided data from the linked list.

        Parameters:
        data: The data of the node to remove.
        
        Returns:
        None
        """
        if not self.head: # If the linked list is empty
            return

        if self.head.data == data: # If the head node has the provided data
            self.head = self.head.next
            return

        current = self.head
        while current.next:
            if current.next.data == data: # If the next node has the provided data
                current.next = current.next.next # Skip the next node
                return
            current = current.next

    def is_empty(self):
        """
        Checks if the linked list is empty.

        Returns:
        bool: True if the linked list is empty, False otherwise.
        """
        return self.head is None
    
    def remove_duplicates(self):
        """
        Removes duplicate elements from the linked list.
        """
        if self.is_empty(): # If the linked list is empty
            return

        current = self.head
        visited = set([current.data]) # Keep track of visited elements using a set
        while current.next:
            if current.next.data in visited:
                current.next = current.next.next # Skip duplicate element
            else:
                visited.add(current.next.data)
                current = current.next

    
    
class Stack:
    def __init__(self):
        """
        Initializes a new instance of the Stack class.
        """
        self.linked_list = LinkedList()

    def append(self, data):
        """
        Pushes a new element onto the stack.

        Parameters:
        data: The data to be pushed onto the stack.
        
        Returns:
        None
        """
        self.linked_list.append(data)

    def remove(self):
        """
        Removes and returns the top element from the stack.

        Returns:
        The top element from the stack.
        """
        if self.is_empty():
            raise IndexError("Stack is empty")
        else:
            data = self.linked_list.head.data
            self.linked_list.remove(data)
            return data
        
    def extend(self, my_list):
        """
        Extends the current stack by pushing all elements from another stack.

        Parameters:
        my_list: The stack to extend the current stack with.
        
        Returns:
        None
        """
        current = my_list.linked_list.head
        while current:
            self.append(current.data)
            current = current.next
            
    def is_empty(self):
        """
        Checks if the stack is empty.

        Returns:
        bool: True if the stack is empty, False otherwise.
        """
        return self.linked_list.is_empty()
```

# Book Class

```python

list_id = LinkedList()

class Book:
  
   # Represents a book with its metadata, such as title, authors, summary, language, publisher, page count, publication year, price, likes, dislikes, and readers.
   

    def __init__(self, title, authors, summary, language, publisher, page_count, publication_year):
        """
        Initializes a new Book object with the given metadata.

        Parameters:
        title (str): The title of the book.
        authors (list): A list of authors of the book.
        summary (str): A summary or description of the book.
        language (str): The language in which the book is written.
        publisher (str): The publisher of the book.
        page_count (str): The total number of pages in the book.
        publication_year (str): The year of publication of the book.

        """
        
        # Generate a random ID for the book
        num = random.randint(1, 1000000)
        count = 0
        current = list_id.head
        
        # Check if the generated ID already exists in the linked list:
        while current:
            if current.data == num:
                count = 1
            current = current.next
        # If the generated ID already exists, generate a new one until a unique ID is found :
        while count == 1:
            num = random.randint(1, 1000000)
            count = 0
            current = list_id.head
            while current:
                if current.data == num:
                    count = 1
                current = current.next
        
        self.id = num
        list_id.append(num)
        
        # Initialize the other attributes and data for the Book object :
        self.feedbacks = LinkedList()
        self.title = title
        self.authors = authors
        self.summary = summary
        self.language = language
        self.publisher = publisher
        self.page_count = page_count
        self.publication_year = publication_year
        self.price = round(random.uniform(80, 1000), 2)
        self.likes = random.randint(20, 50000)
        self.dislikes = random.randint(20, 50000)
        self.readers = LinkedList()
    
    def get_id(self):
        """
        Returns the ID of the book.

        Returns:
        int: The ID of the book.
        """
        return self.id
    
    def add_reader(self, user):
        """
        Adds a reader to the book's list of readers and adds the book to the user's readbook list.

        Parameters:
        user (User): The user object representing the reader.

        Returns:
        None
        """
        self.readers.append(user)
        user.readbook.append(self)
    
   


    def print_info(self):
            """
            Print the information about the book.

            Displays the attributes and their corresponding values of the book in a tabular format.

            Attributes:
                - Authors: List of authors of the book.
                - Summary: Summary of the book.
                - Language: Language of the book.
                - Publisher: Publisher of the book.
                - Page Count: Number of pages in the book.
                - Publication Year: Year of publication of the book.
                - Likes: Number of likes received for the book.
                - Dislikes: Number of dislikes received for the book.
                - Price: Price of the book.
                - Feedbacks :List of feedbacks for the book.

            """
            a = self.title
            a = a.upper()
            print(f"TITLE: {a}:")
            pd.set_option('display.max_colwidth', None)
            df = pd.DataFrame(columns=["Attribute", "Value"])
            df.loc[0] = ["Authors", ", ".join(self.authors)]
            df.loc[1] = ["Summary", self.summary]
            df.loc[2] = ["Language", self.language]
            df.loc[3] = ["Publisher", self.publisher]
            df.loc[4] = ["Page Count", self.page_count]
            df.loc[5] = ["Publication Year", self.publication_year]
            df.loc[6] = ["Likes", self.likes]
            df.loc[7] = ["Dislikes", self.dislikes]
            df.loc[8] = ["Price", self.price]
            
            feedbacks=[] # List to store the feedbacks for the book
            head = self.feedbacks.head
            while head :
                feedbacks.append(head.data)
                head=head.next
            df.loc[9] = ["Feedbacks", ", ".join(str(feedback) for feedback in feedbacks)]
            display(df) # Display the DataFrame
```

## Book Functions: Extraction, Filtering and Sorting

```python

def insertion_sort_publication_year(l):
    """
    Sorts a linked list of Book objects based on their publication year in ascending order using the insertion sort algorithm.

    Parameters:
    l (LinkedList): The linked list to sort.
    
    Returns:
    None
    
    """
    if not l.head or not l.head.next:
        return

    sorted_list = LinkedList() # Create a new linked list to store the sorted elements
    current = l.head

    while current:
        next_node = current.next
        insert_node(sorted_list, current) # Insert the current node into the sorted list
        current = next_node

    l.head = sorted_list.head # Update the head of the original linked list with the head of the sorted list

def insert_node(l, node):
    """
    Inserts a node with Book object data into the appropriate position in a linked list sorted by publication year.

    Parameters:
    l (LinkedList): The linked list to insert the node into.
    node (Node): The node to insert.
    
    Returns:
    None
    
    """
    if not l.head or node.data.publication_year <= l.head.data.publication_year:
         # If the linked list is empty or the node's publication year is smaller than or equal to the head's publication year
         # Set the node as the new head and update its next reference to the current head
        node.next = l.head
        l.head = node
        return

    current = l.head

    while current.next and current.next.data.publication_year < node.data.publication_year:
        # Iterate through the linked list until reaching the correct position to insert the node
        current = current.next

    # Insert the node by updating the next references : 
    node.next = current.next
    current.next = node

def extract_books_metadata_from_api(query, max_results=10):
    """
    Extracts book metadata from the Google Books API based on the given query.

    Parameters:
    query (str): The search query.
    max_results (int): The maximum number of results to retrieve (default: 10).

    Returns:
    LinkedList: A linked list containing Book objects with the retrieved metadata.
    """
    base_url = 'https://www.googleapis.com/books/v1/volumes'
    params = {
        'q': query,
        'maxResults': max_results, # The maximum number of results to retrieve
        'printType': 'books', # Specify that we want to retrieve books
        'fields': 'items(volumeInfo/title,volumeInfo/authors,volumeInfo/description,volumeInfo/language,volumeInfo/publisher,volumeInfo/pageCount,volumeInfo/publishedDate)'
    }

    response = requests.get(base_url, params=params) # Send a GET request to the API

    if response.status_code == 200: # If the response is successful
        data = response.json() # Extract the JSON data from the response
        books = LinkedList() # Create a linked list to store the retrieved books
        for item in data.get('items', []):
            volume_info = item.get('volumeInfo', {})
            # Create a new Book object using the extracted metadata : 
            book = Book(
                title=volume_info.get('title', ''),
                authors=volume_info.get('authors', []),
                summary=volume_info.get('description', ''),
                language=volume_info.get('language', ''),
                publisher=volume_info.get('publisher', ''),
                page_count=volume_info.get('pageCount', ''),
                publication_year=volume_info.get('publishedDate', '')[:4]
            )
            books.append(book)
        return books
    else:
        # If the response status code is not 200, return an empty linked list :
        return LinkedList()


def filter_existing_books(query, max_results=20):
    """
    Filters existing books based on the given query and compares them to the books in the list 'our_books'
    to avoid creating duplicate Book instances with different attribute values. The filtered books are then sorted
    in ascending order based on their publication year.

    Parameters:
    query (str): The search query.
    max_results (int): The maximum number of results to retrieve (default: 20).

    Returns:
    LinkedList: A linked list containing filtered and sorted Book objects.
    """
    # Extract book metadata from the Google Books API: 
    book = extract_books_metadata_from_api(query, max_results)
    
    # Compare and update book attributes from 'our_books' list :
    current = book.head
    while current:
        current1 = our_books.head
        while current1:
            if current.data.title == current1.data.title:
                current.data = current1.data
            current1 = current1.next
        current = current.next
    
    # Filter out books with missing or empty attributes : 
    book_metadata = book.head
    while book_metadata:
        if (
            book_metadata.data.title == ""
            or book_metadata.data.authors == []
            or book_metadata.data.summary == ""
            or book_metadata.data.language == ""
            or book_metadata.data.publisher == ""
            or book_metadata.data.page_count == ""
            or book_metadata.data.publication_year == ""
        ):
            book.remove(book_metadata.data)
        book_metadata = book_metadata.next
    
    # Sort the filtered books by publication year :
    insertion_sort_publication_year(book)
    return book
```

# User Class

```python
class User:
    """
    Represents a user with their name and book-related attributes, such as their liked books, disliked books, read books, and search history.
    """

    def __init__(self, name: str):
        """
        Initializes a new User object with the given name.

        Parameters:
        name (str): The name of the user.
        """
        self.name = name
        self.readbook = LinkedList()
        self.likebook = LinkedList()
        self.dislikebook = LinkedList()
        self.historic = Stack()

    def set_name(self, name: str):
        """
        Sets the name of the user.

        Parameters:
        name (str): The name of the user.

        Returns:
        None
        """
        self.name = name

    def like(self, book_id):
        """
        Adds a like to the book with the specified book_id and appends it to the user's liked books.

        Parameters:
        book_id: The ID of the book to be liked.

        Returns:
        None
        """
        current = our_books.head
        while current:
            if current.data.id == book_id:
                current.data.likes += 1
                self.likebook.append(current.data)
            current = current.next

    def dislike(self, book_id):
        """
        Adds a dislike to the book with the specified book_id and appends it to the user's disliked books.

        Parameters:
        book_id: The ID of the book to be disliked.

        Returns:
        None
        """
        current = our_books.head
        while current:
            if current.data.id == book_id:
                current.data.dislikes += 1
                self.dislikebook.append(current.data)
            current = current.next

    def add_feedback_book(self, id_book, feed_back):
        """
        Adds feedback to the book with the specified id_book.

        Parameters:
        id_book: The ID of the book to add feedback to.
        feed_back: The feedback to be added.

        Returns:
        None
        """
        current = our_books.head
        while current:
            if current.data.id == id_book:
                current.data.feedbacks.append(feed_back)
                break
            current = current.next

    def add_book(self, book_id):
        """
        Adds the book with the specified book_id to the user's read books, and adds the user to the attribute readers(LinkedList) of the book.

        Parameters:
        book_id: The ID of the book to be added.

        Returns:
        None
        """
        current = our_books.head
        while current:
            if current.data.id == book_id:
                self.readbook.append(current.data)
                current.data.add_reader(self)
            current = current.next
            
    def remove_book(self, book_id):
        """
        Removes the book with the specified book_id from the user's read books.

        Parameters:
        book_id: The ID of the book to be removed.

        Returns:
        None
        """
        current = our_books.head
        while current:
            if current.data.id == book_id:
                self.readbook.remove(current.data)
            current = current.next

    def search(self, key_word):
        """
        Searches for books based on the specified key_word and appends them to the user's search history.

        Parameters:
        key_word: The keyword to search for books.

        Returns:# can be removed
        LinkedList: The list of books found based on the search.
        """
        # Filter and retrieve books based on the key_word :
        book1_metadata = filter_existing_books(key_word)
        
        book_metadata = book1_metadata.head
        while book_metadata is not None:
            our_books.append(book_metadata.data)
            book_metadata.data.print_info() # Print the information about the book
            book_metadata = book_metadata.next
            
        # Append the key_word to the user's search history :
        self.historic.append(key_word)
        return book1_metadata

    def new_recommender(self):
        """
        Provides book recommendations based on genre for a user without a search history.

        Returns:
        None
        """
        current = genre_book.head
        # Iterate over each genre in the 'genre_book' linked list :
        while current:
            print("GENRE:", current.data[0],"\n") # Print the genre name
            # Filter books based on a randomly selected sub-genre from the current genre : 
            l = filter_existing_books(random.choice(current.data[1]), 5)
            next1 = l.head
            
            # Iterate over the filtered book list and perform operations for each book :
            while next1:
                our_books.append(next1.data)
                next1.data.print_info()
                print("\n")
                next1 = next1.next
                
            current = current.next # Move to the next genre in the 'genre_book' linked list
            print("\n")

    def enter_description(self, texto):
        """
        Searches for books based on the specified description and appends the description to the user's search history.

        Parameters:
        texto: The description to search for books.

        Returns:
        None
        """
        l = filter_existing_books(texto)
        self.historic.append(texto)
        book_metadata = l.head
        while book_metadata:
            book_metadata.data.print_info()
            book_metadata = book_metadata.next

    def historical_recommender(self):
        """
        Provides book recommendations based on the user's historical data.

        Returns:
        LinkedList: The list of recommended books.
        """
        if self.historic:
            # Initialize an empty linked list to store the recommended books :
            my_list = LinkedList()
            
            # Collaborative filtering based on reading history : 
            read_book = self.readbook.head
            while read_book:
                reader = read_book.data.readers.head
                while reader :
                    if reader.data == self:
                        reader = reader.next
                    else:
                        # Add books from the reader's historical data to the recommendation list : 
                        word = reader.data.historic.linked_list.head
                        while word :
                            my_list.extend (filter_existing_books(word.data,2))
                            word = word.next
                         # Add books liked by the reader to the recommendation list :
                        like = reader.data.likebook.head
                        while like :
                            my_list.extend(filter_existing_books(like.data.authors,2))
                            my_list.extend(filter_existing_books(like.data.title,2))
                            like=like.next
                        reader= reader.next
                read_book=read_book.next
                
            # Add books from the user's search history to the recommendation list : 
            current = self.historic.linked_list.head
            while current:
                my_list.extend(filter_existing_books(current.data, 5))
                current = current.next
                
            # liked list:
            current = self.likebook.head
            while current:
                my_list.extend(filter_existing_books(current.data.authors, 5))
                my_list.extend(filter_existing_books(current.data.title, 5))
                current = current.next
                
            # Remove books from the recommendation list that are disliked by the user : 
            current = self.dislikebook.head
            while current:
                cur = my_list.head
                while cur:
                    if current.data.authors == cur.data.authors or current.data.title == cur.data.title:
                        my_list.remove(cur.data)
                    cur = cur.next
                current = current.next
            # Read list:
            current = self.readbook.head
            while current:
                # Recommend books by the same authors :
                my_list.extend(filter_existing_books(current.data.authors, 5))
                # Remove the current book from the recommendation list :
                my_list.remove(current.data)
                # Recommend books with similar titles:
                my_list.extend(filter_existing_books(current.data.title, 5))
                # Remove duplicates of the current book from the recommendation list:
                my_list.remove(current.data)
                my_list.remove(current.data)
                current = current.next
            my_list.remove_duplicates() # Remove duplicate books from the recommendation list
            insertion_sort_publication_year(my_list) # Sort the recommendation list by publication year
            book_metadata = my_list.head
            while book_metadata:
                book_metadata.data.print_info() # Print the information about each recommended book
                book_metadata = book_metadata.next

            return my_list


    def filter_search(self, keyword, page_count=range(0, 10000000), price=range(80, 1001), likes=20, language="fren",
                      pub_year=0):
        """
        Filters search results based on various parameters and appends the filtered books to the user's search history.

        Parameters:
        keyword: The keyword to search for books.
        page_count: The range of page counts to filter the books (default: range(0, 10000000)).
        price: The range of prices to filter the books (default: range(80, 1001)).
        likes: The minimum number of likes to filter the books (default: 20).
        language: The language to filter the books (default: "fren").
        pub_year: The minimum publication year to filter the books (default: 0).

        Returns:
        None
        """
        try:
        # Convert input values to appropriate types and handle potential errors
            page_count = eval(page_count)
            price = eval(price)
            likes = int(likes)
            pub_year = int(pub_year)
        except :
            print("INVALID INPUT")
            return
        
        book = filter_existing_books(keyword)
        book_metadata = book.head
        while book_metadata:
            # Filter books based on the specified parameters : 
            if (
                int(book_metadata.data.price) not in price
                or book_metadata.data.language not in language
                or book_metadata.data.likes <= likes
                or int(book_metadata.data.page_count) not in page_count
                or int(book_metadata.data.publication_year) <= pub_year
            ):
                book.remove(book_metadata.data)
            book_metadata = book_metadata.next
        book1 = book.head
        while book1 is not None:
            our_books.append(book1.data)
            book1.data.print_info()
            book1 = book1.next
        self.historic.append(keyword)

    def reset(self):
        """
        Resets the user's read, liked, disliked books, and search history.

        Returns:
        None
        """
        self.readbook = LinkedList()
        self.likebook = LinkedList()
        self.dislikebook = LinkedList()
        self.historic = LinkedList()
        
    def guide(self):
        """
        Provides a guide to using ReadQuest, introducing the user to the available methods and their functionalities.

        Returns:
        None
        """
        print("Hello",self.name,"!\n")
        print("Welcome to ReadQuest, your ultimate reading companion.Our innovative recommendation system will be your trusted guide, unlocking a world of literary wonders personalized just for you.")
        print ("To make the most of your ReadQuest experience, we have equipped you with an array of powerful methods. Each method is designed to enhance your reading adventure and bring you closer to your literary desires. Let's take a closer look at what each method has to offer:")
        infos = '''Like(book_id):
By using this method, you can express your appreciation for a particular book. Simply provide the book ID, and ReadQuest will register your like, allowing us to fine-tune future recommendations based on your preferences.

Dislike(book_id):
If a book doesn't resonate with you, don't worry! Use the Dislike method with the book ID, and ReadQuest will take note, ensuring that similar books won't appear in your recommendations.

Add_book(book_id):
Found a book that captures your interest? Add it to your reading list using this method. ReadQuest will keep track of the books you're currently reading, making it easier for you to manage and navigate your literary adventures.

Remove_book(book_id):
If you've completed a book or simply wish to remove it from your reading list, the Remove_book method is here to help. Just provide the book ID, and ReadQuest will bid farewell to that book from your list.

Search(key_word):
Discover new books that align with your interests through the Search method. Enter a keyword, and ReadQuest will fetch a collection of books that match your search. Dive into their information and explore intriguing summaries to find your next literary gem.

Filter_search(keyword, page_count, price, likes, language, pub_year):
Refine your search even further with the Filter_search method. Specify various parameters such as page count, price range, likes, language, and publication year to narrow down your book recommendations according to your preferences.

New_recommender():
Looking for a fresh wave of inspiration? The New_recommender method provides a selection of books based on different genres. Let ReadQuest surprise you with handpicked recommendations and introduce you to new authors and captivating stories.

Enter_description(texto):
If you have a specific description or topic in mind, the Enter_description method will fetch books that match your description. Immerse yourself in a world of literary possibilities and explore books that align with your interests.

Historical_recommender():
Take a deep dive into personalized recommendations with the Historical_recommender method. ReadQuest analyzes your reading history, likes, dislikes, and search patterns to curate a list of books tailored specifically for you. Prepare to be amazed by the books that resonate with your unique reading preferences.

Reset():
Need a fresh start? The Reset method wipes your reading history, liked and disliked books, and search records, giving you a clean slate to embark on new literary adventures.'''
        print("\n"+infos)
        print("\nWith these powerful methods at your disposal, ReadQuest will unlock a world of literary wonders tailored exclusively for you. Immerse yourself in captivating stories, explore new genres, and connect with your fellow readers on this magical journey.")
        print("\nHappy reading!")
```
